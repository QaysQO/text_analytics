```python
# Importing the regular expression module
import re
# Importing the JSON module
import json
# Importing the requests module for making HTTP requests
import requests
# Importing datetime module from datetime for handling timestamps
from datetime import datetime 
#import csv to extract file
import csv

#get url site
url = 'https://shopee.com.my/(USED)-ASUS-ROG-MSI-GIGABYTE-ZOTAC-PALIT-EVGA-GRAPHIC-CARD-GTX-RTX-2060SUPER-3060-3060TI-3070-3070TI-3080-3080-TI-3090-i.122126415.20359007421?sp_atk=5e4b6cfc-0064-44d0-812e-968edb8632b1&xptdk=5e4b6cfc-0064-44d0-812e-968edb8632b1'

# Searching for a pattern in the URL
r = re.search(r'i\.(\d+)\.(\d+)', url)  

# Extracting shop_id and item_id 
if r:  
    shop_id, item_id = r.group(1), r.group(2)  
else:  
    print("Invalid URL")  
    exit()  

# Initializing a counter for pages
count_page = 0  
# Initializing an offset for pagination
offset = 0  
# Filename for CSV export
filename = 'Lab1_IS01081503_IS01081203.csv'

# Open the CSV file in write mode
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    # Field names for the CSV
    fieldnames = ['author', 'date', 'comment']  
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
     # Write the header row
    writer.writeheader()
    #loop for process
    while True:  
        # Constructing the URL for fetching ratings
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'  
        # Sending a GET request to the ratings URL and converting the response to JSON
        data = requests.get(ratings_url).json() 
    
        #loop for each data
        for rating in data['data']['ratings']: 
            # Printing the author's username
            author = rating['author_username']
            # Printing and converting timestamp to a readable date format
            date_review = datetime.fromtimestamp(rating['ctime']).strftime('%Y-%m-%d')    
            # Printing the rating comment
            comment = rating['comment']  
            
            writer.writerow({'author': author, 'date': date_review, 'comment': comment})

        # Incrementing the page counter
        count_page += 1  
    
        # Checking if 5 pages have been processed
        if count_page == 5:  
            break  

        # Checking if there are less than 20 ratings in the response
        if len(data['data']['ratings']) <= 20:  
            break  

        # Incrementing the offset for pagination
        offset += 20  # Incrementing the offset for pagination
        
    print(f"Data exported to {filename}")
```
